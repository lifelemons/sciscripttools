```python
class ExponentialSmoothing:

    s_old = 0.0
    alpha = 0.5;
    
    def __call__(self, x):
        s = exponential_smoothing(x, self.s_old, self.alpha);
        self.s_old = s
        return s
```

```python
def exponential_smoothing(x, s_old, alpha):
    """
    Compute the exponentially smoothed value

    Parameters:
    -----------
    x: float
        the new value
    s_old: float
        the previous moving average value 
    alpha: float
        Smoothing constant
    """
    return alpha * x + (1 - alpha) * s_old


def smoothing_constant(n):
    """
    n: is the window size, i.e. the number of samples
    """
    return 2.0 / (n + 1)
```

```python
def exponential_smoothed(x, alpha):
    smoothed_x = x[0]
    smoothed_data = []
    for x_i in x:
        smoothed_x = exponential_smoothing(x_i, smoothed_x, alpha)
        smoothed_data.append(smoothed_x)

    return np.array(smoothed_data)
```

```python
def exponential_smoothed_zero_start(x, alpha_stable, alpha_unstable):
    smoothed_x = 0
    smoothed_data = []
    #input_alpha = alpha
    for x_i in x:

        if (np.abs(x_i - smoothed_x) < smoothed_x * 0.05):
            alpha = alpha_stable
        else:
            alpha = alpha_unstable

        smoothed_x = exponential_smoothing(x_i, smoothed_x, alpha)
        smoothed_data.append(smoothed_x)

    return np.array(smoothed_data)
```

```python
## Generate random data

mean_value = 500   # The mean around which to center the distribution
standard_deviation = mean_value * 0.1  # Standard deviation of the distribution
num_samples = 100  # Number of samples to generate

# Generate the random distribution
data = np.random.normal(loc=mean_value, scale=standard_deviation, size=num_samples)
```

```python
def create_time_series(n, interval):
    end_time = 0 + interval * (n - 1)
    return np.linspace(0, end_time, n)
```
